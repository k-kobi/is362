Let's tidy up this "wide" dataset of yearly exam scores!

```python
import pandas as pd

# Read the CSV file into a pandas DF
df_wide = pd.read_csv('/home/kobi/is362/Project 2/Exam Scores.csv')

# Print the df
print("Dataset pre tidying:")
display(df_wide)
```

```python
# convert to long
df_tidy = pd.melt(df_wide, id_vars=['Name'], var_name='Year', value_name='Score')

# clean the year by removing "Score_"
df_tidy['Year'] = df_tidy['Year'].str.replace('Score_', '')

# Display the tidy dataset (post-cleaning)
print("Tidy Dataset:")
display(df_tidy)
```

We now have a tidy data set!

1) Since I made this data set myself, it did not really take much time to find... but only a few minutes to create and clean!

2) By preparing the tidy dataset, I am now able to run easy analyses with the aggregate data! See below for my average exam scores for this class!

3) There wasn't much to keep in mind here to maintain the data structure other than maintaining the relationship between the variables.

```python
# Average exam scores for this class:

average_score = df_tidy['Score'].mean()
print(f"Overall Average Score: {average_score}")

# Average exam score grouped by year:

average_score_by_year = df_tidy.groupby('Year')['Score'].mean()
print("Average Score by Year:")
print(average_score_by_year)
```

As we can see, we were easily able to run analyses, even an easy analysis of grouping by year, which the original dataset aimed to do in a very "wide" way!
